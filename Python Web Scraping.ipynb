````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
def fetch_movie_urls(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return "Failed to retrieve the page, status code:", response.status_code
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup


url = 'https://www.justwatch.com/in/movies?release_year_from=2000'
soup=fetch_movie_urls(url)
print(soup.prettify())

## Hint : Use the following code to extract the film urls
movie_links = soup.find_all('a', href=True)
movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

````

## **Fetching Movie URL's**

````python
# Write Your Code here
url_list=[] #Empty list to store URL
for x in movie_urls:
  url_list.append('https://www.justwatch.com'+x)
print(url_list)
````

## **Scrapping Movie Title**

````python

Title =[] #for creating empty list of title
for url in url_list:
  #] Extract the movie name using regex
  movie_name = re.search(r'movie/([^/]+)', url).group(1)

  # Replace hyphens with spaces to make it readable
  movie_name = movie_name.replace('-', ' ')

  Title.append(movie_name)
print(Title)
````

## **Scrapping release Year**

````python
# Write Your Code here
release_year = []
for url in url_list:
 soup=fetch_movie_urls(url)
 #print(soup.prettify())

# Search for the element containing the release year
 json_ld_script = soup.find("script", {"type": "application/ld+json"})

 if json_ld_script:
    # Extract the text content of the script tag
    json_text = json_ld_script.get_text()

    # Use regex to find a 4-digit year (usually release year)
    match = re.search(r'"(?:dateCreated|uploadDate|name|@id)":"?(\d{4})', json_text)

    if match:
      release_year.append(match.group(1))
        #print("Release Year:", match.group(1))
    else:
        print("Release Year not found in JSON-LD.")
print(release_year)
````

## **Scrapping Genres**

````python
# Write Your Code here
Genres = []
for url in url_list:
  soup=fetch_movie_urls(url)
  genres_div = soup.find("h3", string="Genres")  # Find the <h3> with text "Genres"
  if genres_div:
    genres_value = genres_div.find_next_sibling("div")  # Get the next <div> with genre values
    if genres_value:
      genres = genres_value.get_text(strip=True)  # Extract text
    else:
        genres = "N/A"
  else:
      genres = "N/A"
  Genres.append(genres)


print(Genres)

````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
IMDBRating = []
for url in url_list:
  soup=fetch_movie_urls(url)
  imdb_span = soup.find("span", class_="imdb-score")  # Find the <h3> with text "imdb-score"
  imdb_rating = imdb_span.get_text(strip=True) if imdb_span else "N/A"  # Extract text
  IMDBRating.append(imdb_rating)

print(IMDBRating)
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
Runtime = []
for url in url_list:
  soup=fetch_movie_urls(url)
  Ruttime_div = soup.find("h3", string="Runtime")  # Find the <h3> with text "Runtime"
  if Ruttime_div:
    Ruttime_div = Ruttime_div.find_next_sibling("div")  # Get the next <div> with Runtime values
    if Ruttime_div:
      Ruttime_div = Ruttime_div.get_text(strip=True)  # Extract text
    else:
        Ruttime_div = "N/A"
  else:
      Ruttime_div = "N/A"
  Runtime.append(Ruttime_div)


print(Runtime)

````

## **Scrapping Age Rating**

````python
# Write Your Code here
Age_rating = []
for url in url_list:
  soup=fetch_movie_urls(url)
  Age = soup.find("h3", string="Age rating")  # Find the <h3> with text "Age rating"
  if Age:
    Age = Age.find_next_sibling("div",class_="poster-detail-infos__value")  # Get the next <div> with Age rating values
    if Age:
      Age = Age.get_text(strip=True)  # Extract text
    else:
        Age = "N/A"
  else:
      Age = "N/A"
  Age_rating.append(Age)


print(Age_rating)
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
Production_Country = []
for url in url_list:
  soup=fetch_movie_urls(url)
  ProductionCountry = soup.find("h3", string="Production country")  # Find the <h3> with text "Production country"
  if ProductionCountry:
    ProductionCountry = ProductionCountry.find_next_sibling("div",class_="poster-detail-infos__value")  # Get the next <div> with Production country values
    if ProductionCountry:
      ProductionCountry = ProductionCountry.get_text(strip=True)  # Extract text
    else:
        ProductionCountry = "N/A"
  else:
      ProductionCountry = "N/A"
  Production_Country.append(ProductionCountry)


print(Production_Country)
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
Streamingservices = []

for url in url_list:
    soup = fetch_movie_urls(url)
    service_images = soup.find_all("link", {"as": "image"})
    services = set()  # Using a set to avoid duplicates

    for img in service_images:
        href = img.get("href")
        if href:
            if "netflix" in href:
                services.add("Netflix")
            elif "amazonprimevideo" in href:
                services.add("Prime Video")
            elif "zee5" in href:
                services.add("ZEE5")
            elif "itunes" in href:
                services.add("Apple TV (iTunes)")
            elif "amazon" in href:
                services.add("Amazon")

    Streamingservices.append(services)

print(Streamingservices)
````

````python
# Write Your Code here
data = pd.concat([MovieData,tvshowData],ignore_index=True) # Append both Data Frame into Single DataFrame
data.info()

# Convert IMDB rating into Float data type
data['IMDBRating_Cleaned'] = data['IMDBRating'].replace("N/A",0) #For Movie IBDB replacing N/A Value with 0
data['IMDBRating_Cleaned'] = data['IMDBRating_Cleaned'].str.split(" ").str[0].astype(float)  # Extracting first number in rating
data["ReleaseYear"] = data["ReleaseYear"].astype(int)

#For Only include movies and TV shows released in the last 2 years (from the current date).
from datetime import datetime
currentyear = datetime.now().year #Find out Current year
Movieshow_last2Yr = data[data["ReleaseYear"] >= (currentyear-2)]
Movieshow_last2Yr.info()

# For Only include movies and TV shows with an IMDb rating of 7 or higher.
MovieShow_Highrating = data[data["IMDBRating_Cleaned"] >= 7]
MovieShow_Highrating.info()
````
